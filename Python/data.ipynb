```python
columns_needed = [
    'loan_amnt',
    'funded_amnt',
    'funded_amnt_inv',
    'term',
    'int_rate',
    'installment',
    'grade',
    'sub_grade',
    'emp_length',
    'home_ownership',
    'annual_inc',
    'verification_status',
    'issue_d',
    'loan_status',
    'purpose',
    'addr_state',
    'dti',
    'fico_range_low',
    'fico_range_high',
    'total_pymnt',
    'total_rec_prncp',
    'total_rec_int'
]

print(data1[columns_needed].head())
```

```python
loan_data = data1[columns_needed].copy()
```

```python
loan_data.isnull().sum()
```

```python
missing_percent = loan_data.isnull().mean() * 100

cols_to_drop = missing_percent[missing_percent > 50].index

loan_data.drop(columns=cols_to_drop, inplace=True)

print("Dropped columns:", cols_to_drop.tolist())
```

```python
num_cols = loan_data.select_dtypes(include=['int64', 'float64']).columns

loan_data[num_cols] = loan_data[num_cols].fillna(loan_data[num_cols].median())
```

```python
cat_cols = loan_data.select_dtypes(include='object').columns

loan_data[cat_cols] = loan_data[cat_cols].fillna("Unknown")
```

```python
loan_data['issue_d'] = pd.to_datetime(
    loan_data['issue_d'],
    errors='coerce'
)
```

```python
loan_data['Year'] = loan_data['issue_d'].dt.year
loan_data['Month'] = loan_data['issue_d'].dt.month_name()
```

```python
good_status = ['Fully Paid', 'Current']
bad_status = ['Charged Off', 'Default', 'Late (31-120 days)']

loan_data['Loan_Category'] = 'Other'

loan_data.loc[
    loan_data['loan_status'].isin(good_status),
    'Loan_Category'
] = 'Good Loan'

loan_data.loc[
    loan_data['loan_status'].isin(bad_status),
    'Loan_Category'
] = 'Bad Loan'
```

```python
loan_data['Loan_Category'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%',
    figsize=(7,7)
)

plt.ylabel("")
plt.title("Good vs Bad Loans")
plt.show()
```
